```python
# get medicare drug utilization data from the 2017 set 
medicare2017 = pd.read_csv('https://practicum-mitre.s3.amazonaws.com/Medicare_Provider_Utilization_and_Payment_Data__2017_Part_D_Prescriber.csv')
medicare2017.head()
```

```python
# read in drug NDC code information to use later to filter medicare dataset
ndc = pd.read_excel('HEDIS-2019-NDC-MLD-Directory-Complete-Workbook-FINAL-11-1-2018-2.xlsx',
                         'Medications List to NDC Codes')
ndc.head()
```

```python
# creating a df with only opioids and we capture all opioids by seeng where MED (Morphine Equivalent Dose) is not nan
opioiddf = ndc[ndc['MED Conversion Factor'].notnull()].copy()
opioiddf['Drug Name'] = opioiddf['Drug Name'].str.upper()
opioiddf.head()
```

```python
#Check the opioids in the df
opioiddf['Drug Name'].unique()
```

```python
# make the opioid drug column into a list 
opioidlist = opioiddf['Drug Name'].unique().tolist()
# join together list of opioids to create a pattern for comparison
opioidlist2015 = '|'.join(opioidlist)
opioidlist2016 = '|'.join(opioidlist)
opioidlist2017 = '|'.join(opioidlist)
```

```python
# use str.contains to match the two patterns from the two datasets
opioidlist2015_ = medicare2015[(medicare2015['drug_name'].str.contains(opioidlist2015)) | (medicare2015['generic_name'].str.contains(opioidlist2015))]
opioidlist2016_ = medicare2016[(medicare2016['drug_name'].str.contains(opioidlist2016)) | (medicare2016['generic_name'].str.contains(opioidlist2016))]
opioidlist2017_ = medicare2017[(medicare2017['drug_name'].str.contains(opioidlist2017)) | (medicare2017['generic_name'].str.contains(opioidlist2017))]
```

```python
#List the opioids
opioidlist2015
```

```python
opioidlist2016
```

```python
opioidlist2017
```

```python
opioidlist_final = opioidlist2015
opioidlist_final
```

```python
#Function to combine into one dataframe
def opioid_df(df):
    df = df[df['generic_name'].str.contains(opioidlist_final)].copy()
    return df
```

```python
#Create a singular opioid dataframe for medicare for each year
opioid2015 = opioid_df(medicare2017)
opioid2016 = opioid_df(medicare2016)
opioid2017 = opioid_df(medicare2017)
```

```python
#See unique drug names
opioid2015.drug_name.unique()
opioid2016.drug_name.unique()
opioid2017.drug_name.unique()
```

```python
# There were drugs in the combined dataset that weren't explicitly opioids, so we remove them
otherdrugs = ['COMBIVENT RESPIMAT', 'SPIRIVA RESPIMAT', 'ATROVENT', 'STIOLTO RESPIMAT', 'IPRATROPIUM BROMIDE', 'SPIRIVA', 'IPRATROPIUM-ALBUTEROL']

opioid2015 = opioid2015[~opioid2015.drug_name.isin(otherdrugs)]
opioid2016 = opioid2016[~opioid2016.drug_name.isin(otherdrugs)]
opioid2017 = opioid2017[~opioid2017.drug_name.isin(otherdrugs)]
```

```python
#Checking to see if the irrelevant medications were removed
opioid2015.drug_name.unique()
opioid2016.drug_name.unique()
opioid2017.drug_name.unique()
```

```python
#Setting the years
opioid2015['year'] = 2015
opioid2016['year'] = 2016
opioid2017['year'] = 2017
```
